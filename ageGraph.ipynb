```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
path = Path("Resources/OD_data.csv")
```

```python
opioidsDf = pd.read_csv(path)
opioidsDf.head()
```

```python
AnyOpioidDf = opioidsDf[opioidsDf['AnyOpioid'] == 'Y']
AnyOpioidDf.head()
```

```python
Data_df = opioidsDf['Age']
Data_df.head()
```

```python
opioidsDf['bins'] = pd.cut(opioidsDf['Age'], bins=[0,25,37,49,65,120], labels=["0-25","26-37","38-49", "50-61", "62+"])
df = opioidsDf.groupby('bins').size()
df
```

```python
df.plot(kind='pie', autopct='%1.1f%%',radius = .97, title='Age of Any Opioid Death in Connecticut', ylabel='Age Range')
plt.savefig('Age Of Any Opioid Death in Connecticut.png')
plt.show()
```
